```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
WASDE = pd.read_csv("WASDE.csv")
```

```python
print(WASDE.head())
```

```python
fig, axes = plt.subplots(3, 1, figsize=(8, 12))

axes[0].plot(WASDE["year"], WASDE["corn_price"], color="blue")
axes[0].set_title("Corn Price Over Time")
axes[0].set_ylabel("USD/bushel")

axes[1].plot(WASDE["year"], WASDE["total_supply"], color="green")
axes[1].set_title("Total Corn Supply Over Time")
axes[1].set_ylabel("Million bushels")

axes[2].plot(WASDE["year"], WASDE["total_use"], color="red")
axes[2].set_title("Total Corn Demand (Use) Over Time")
axes[2].set_ylabel("Million bushels")
axes[2].set_xlabel("Year")
```

```python
plt.tight_layout()
plt.savefig("Corn_Trends_Combined.png", dpi=300)
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(WASDE["year"], WASDE["total_supply"], color="green", label="Total Supply", linewidth=2)
ax1.plot(WASDE["year"], WASDE["total_use"], color="red", label="Total Demand", linewidth=2)

ax1.set_title("Corn Supply vs Demand (V Graph)")
ax1.set_xlabel("Year")
ax1.set_ylabel("Million bushels")
ax1.legend()

plt.tight_layout()
plt.savefig("Corn_Supply_Demand_VGraph.png", dpi=300)
plt.show()
```
